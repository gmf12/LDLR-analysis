```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
PCA_full=pd.read_csv("//Users//glauciomonteiroferreira//OneDrive//LDLr//MD-results/PCA//LDLr_PCA_merged_full_PCA_proj.csv")
Apo=PCA_full.loc[1:100:1]
Cys184Tyr=PCA_full.loc[101:201:1]
Gly373Asp=PCA_full.loc[202:302:1]
Ala431Thr=PCA_full.loc[303:403:1]
Ile488Thr=PCA_full.loc[404:504:1]
Gly549Asn=PCA_full.loc[505:606:1]
Gly592Glu=PCA_full.loc[607:707:1]


plot = sns.jointplot(x="PC_1", y="PC_2", data= Apo, color = "silver", kind="kde", fill=True)


plot.ax_marg_x.set_xlim(-250, 200)
plot.ax_marg_y.set_ylim(-200, 250)

flatui = ["#4E8AA3", "#67C7F0","#F0CB5B","#582EA3","#894FF0","#A36258","#F0D83A","#7DF0D4","#1A5CA3","#F044BD","#59A39B"]


fig1 = plt.gcf()
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlesize'] = 24
plt.xlabel("PC1 (19.5%)")
plt.ylabel("PC2 (17.8%)")
plt.show()
plt.draw()
fig1.savefig('Apo.pdf', dpi=600)
```
